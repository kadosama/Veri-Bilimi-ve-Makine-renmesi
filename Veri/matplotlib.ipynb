```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline #sürekli plt.show dememizi azaltır işlem kolaylığı için yazdık
```

```python
yaslistesi=["10","20","30","40","50"]
kilolistesi=["5","15","25","35","45"]
```

```python
numpyas=np.array(yaslistesi)
numpykilo=np.array(kilolistesi)
```

```python
plt.plot(numpyas,numpykilo,"r")#sondaki harf(Renkler),ilk isim x 2.isim y
```

```python
plt.plot(numpyas,numpykilo,"r")
plt.xlabel("yas")
plt.ylabel("kilo")
plt.title("grafik")
plt.show()#bunu yaz yada yazma farketmez ilkinde onun için kodumuzu yazdık
```

```python
plt.plot(numpyas,numpykilo,"r--")
```

```python
plt.plot(numpyas,numpykilo,"r*-")#her yıldız bi veri
```

```python
plt.subplot(1,2,1)#1 sıra 2 kolon olcak 1.çizicem(bunda axisi olan bi grafiğe çizdik)
plt.plot(numpyas,numpykilo,"r*-")
plt.subplot(1,2,2)
plt.plot(numpyas,numpykilo,"r--");
```

```python
benimfigur=plt.figure()#boş bi graph oluşturdu
figureaxes=benimfigur.add_axes([0.8,0.8,0.6,0.6])#son 2 width height ilk 2 si ise graphin nerde başlacğını sol ve alt köşesinin mesela içiçe grafiklerde kesişmeme olayı kötü görüntü olmama olayı olması için kullanılır
figureaxes.plot(numpyas,numpykilo,"r")
figureaxes.set_xlabel("x")
figureaxes.set_ylabel("y")
figureaxes.set_title("graph")
```

```python
(figur,eksen)=plt.subplots()
print(figur)
print(eksen)
(eksen.plot(numpyas,numpykilo,"r"))
```

```python
(figur,eksen)=plt.subplots(nrows=1,ncols=2)
type(eksen)
```

```python
(figur,eksens)=plt.subplots(nrows=1,ncols=2)
type(eksens)
for a in eksens:
    a.plot(numpyas,numpykilo,"r")
```

```python
benimfigur=plt.figure()#figsize=(3,3) widh height olarak ölçeklendirmesi dpi=150 çözünürlük
figureaxes=benimfigur.add_axes([0.8,0.8,0.6,0.6])
figureaxes.plot(numpykilo,numpyas,label="numpyas**2")
figureaxes.plot(numpykilo,numpyas,label="numpyas**3")
figureaxes.legend(loc=6)#bu etiketin hangi lokasyonda olcağını söler
#farklı olması lazım etikettetki gibi ama yapamadım
```

```python
#kaydetmek için
figureaxes.savefig("benimfigur.png")
figureaxes.savefig("benimfigur.png",dpi=200)
```
